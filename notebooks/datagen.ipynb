```python
import pickle
import numpy as np
import math
import scipy.stats
import seaborn as sns
import matplotlib.pyplot as plt
import copy
%matplotlib inline
sns.set_style("whitegrid")
```

```python
def markovTransition(p,n):
        if np.random.rand() < 0.95 or n==1:
            return p
        else:
            newState = np.random.randint(0, high=n)
            while newState == p:
                newState = np.random.randint(0, high=n)
            p = newState                
        return p
```

```python
def TestData(m0,m1):
    n = 2
    currState = np.random.randint(2)
    states = []
    data = []
    for i in range(500):
        currState = markovTransition(currState,n)
        states.append(currState)
        if currState == 0:
            x = np.random.normal(loc=m0 , scale=1)
        elif currState == 1:
            x = np.random.normal(loc=m1 , scale=1)
        data.append(x)
    return data
```

```python
for d in range(100):
    m0 = 4# np.random.randint(5)
    m1 = -4#np.random.randint(5) * -1
    data = TestData(m0,m1)
    with open('dataset/datamodel-data/{0}-{1}-{2}.txt'.format(m0,m1*-1,d), 'a') as f:
        pickle.dump(data, f)
```
