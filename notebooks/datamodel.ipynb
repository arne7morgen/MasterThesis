## Simulation of Z states

```python
def markovTransition(p):
        if np.random.rand() < 0.95 or n==1:
            return p
        else:
            newState = np.random.randint(0, high=n)
            while newState == p:
                newState = np.random.randint(0, high=n)
            p = newState                
        return p
```

```python
n = 3
currState = np.random.randint(3)
states = []
for i in range(200):
    currState = markovTransition(currState)
    states.append(currState)
plt.scatter(range(200),states)
```

## simulation of data with $\theta$

```python
n = 3
currState = np.random.randint(3)
states = []
data = []
for i in range(200):
    currState = markovTransition(currState)
    states.append(currState)
    if currState == 0:
        x = np.random.normal(loc=-3 , scale=1)
    elif currState == 1:
        x = np.random.normal(loc=4 , scale=0.5)
    elif currState == 2:
        x = np.random.normal(loc=1 , scale=0.4)
    data.append(x)
    
    
plt.figure(figsize=(9,4))
plt.scatter(range(200),data)
plt.xlim( 0, 200 )
plt.title('Data Model')
plt.ylabel('x')
plt.xlabel('t')

plt.show()
plt.plot(range(200), states)
```
